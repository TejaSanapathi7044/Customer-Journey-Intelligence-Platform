```python
products["product_category_name"] = products["product_category_name"].fillna("Unknown")
```

```python
products["product_weight_g"] = products["product_weight_g"].fillna(products["product_weight_g"].median())
```

```python
products["product_length_cm"] = products["product_length_cm"].fillna(products["product_length_cm"].median())

products["product_height_cm"] = products["product_height_cm"].fillna(products["product_height_cm"].median())

products["product_width_cm"] = products["product_width_cm"].fillna(products["product_width_cm"].median())

products["product_name_lenght"] = products["product_name_lenght"].fillna(products["product_name_lenght"].mean())
```

```python
products["product_description_lenght"] = products["product_description_lenght"].fillna(
products["product_description_lenght"].median())

products["product_photos_qty"] = products["product_photos_qty"].fillna(products["product_photos_qty"].median())
```

```python
products.isnull().sum()
```

```python
df=orders.merge(customers,on="customer_id",how="left")
print(df.shape)
```

```python
df=df.merge(orders_payments,on="order_id",how="left")
print(df.shape)
```

```python
df=df.merge(order_items,on="order_id",how="left")
print(df.shape)
```

```python
df=df.merge(products,on="product_id",how="left")
print(df.shape)
```

```python
df=df.merge(orders_reviews,on="order_id",how="left")
print(df.shape)
```

```python
df["order_delivered_customer_date"] = pd.to_datetime(df["order_delivered_customer_date"])
df["order_purchase_timestamp"] = pd.to_datetime(df["order_purchase_timestamp"])
df["delivery_days"]=(df["order_delivered_customer_date"]-df["order_purchase_timestamp"]).dt.days
print(df["delivery_days"])
```

```python
df["order_estimated_delivery_date"] = pd.to_datetime(df["order_estimated_delivery_date"])
df["delay_days"] = (df["order_delivered_customer_date"]- df["order_estimated_delivery_date"]).dt.days
print(df["delay_days"])
```

2.**Order Status Distribution**

```python
df['order_status'].value_counts(normalize=True)*100
```

```python
total_orders = len(df)
canceled_orders = len(df[df["order_status"]== "canceled"])
canceled_orders_percentage = (canceled_orders/total_orders)*100
canceled_orders_percentage
```

```python
delivered_orders = len(df[df["order_status"]=="delivered"])
delivered_orders_percentage = (delivered_orders/total_orders)*100
delivered_orders_percentage
```
